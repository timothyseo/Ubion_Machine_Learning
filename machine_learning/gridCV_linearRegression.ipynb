## GridSearchCV 연습

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=10)

tree = DecisionTreeClassifier()

parameters = {'max_depth':[1,2,3], 'min_samples_split':[2,3]}
```

```python
import pandas as pd

grid = GridSearchCV(tree, param_grid=parameters, cv=3, refit=True)

grid.fit(X_train, y_train)

score = pd.DataFrame(grid.cv_results_)

score[['params', 'mean_test_score', 'rank_test_score', 'split0_test_score', 'split1_test_score', 'split2_test_score']]
print(grid.best_params_)
print(grid.best_score_)
```

```python
from sklearn.metrics import accuracy_score
estimator = grid.best_estimator_

pred = estimator.predict(X_test)
print(accuracy_score(y_test, pred))
```

## Linear Regression - 선형 회귀 연습

```python
import matplotlib.pylab as plt
from sklearn import linear_model

reg = linear_model.LinearRegression()

X = [[174],[152],[138],[128],[186]]
y = [71, 55, 46, 38, 88]

reg.fit(X, y)

print(reg.predict([[165]]))

plt.scatter(X, y, color='black')

pred = reg.predict(X)

plt.plot(X, pred, color='blue', linewidth=5)
plt.show()
```

## Diabetes / Linear Regression practice 

```python
import matplotlib.pylab as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import datasets
from sklearn.model_selection import train_test_split

diabetes=datasets.load_diabetes()


X_train, X_test, y_train, y_test = train_test_split(diabetes.data,diabetes.target, test_size=0.2, random_state=10)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

plt.plot(y_test, y_pred,'.')


# 직선을 그리기 위하여 완벽한 선형 데이터 생성
x=np.linspace(0,330,100)
y=x
plt.plot(x,y)
plt.show()
```
