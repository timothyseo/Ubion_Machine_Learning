```python
df = pd.read_csv('../csv/BND.csv', index_col='Date')
df
```

```python
df = df[~df.isin([np.nan, np.inf, -np.inf]).any(1)]
```

```python
df_1 = df[['Adj Close']]
```

```python
df_1['center'] = df_1['Adj Close'].rolling(20).mean()
```

```python
df_1['ub'] = df_1['center'] + 2 * (df_1['Adj Close'].rolling(20).std())
df_1['lb'] = df_1['center'] - 2 * (df_1['Adj Close'].rolling(20).std())
```

```python
start = '2010-01-01'
df_1 = df_1[start :]
df_1
```

```python
df_1['trade'] = ''
df_1.head(1)
```

```python
for i in df_1.index:
    if df_1.loc[i, 'Adj Close'] > df_1.loc[i, 'ub']:
        df_1.loc[i, 'trade'] = ''
    elif df_1.loc[i, 'Adj Close'] < df_1.loc[i, 'lb']:
        df_1.loc[i,'trade'] = 'buy'
    elif df_1.loc[i, 'Adj Close'] >= df_1.loc[i, 'lb'] and df_1.loc[i, 'Adj Close'] <= df_1.loc[i, 'ub']:
        if df_1.shift(1).loc[i, 'trade'] == 'buy':
            df_1.loc[i, 'trade'] = 'buy'
        else:
            df_1.loc[i, 'trade'] = ''
```

```python
df_1['trade'].value_counts()
```

```python
rtn = 1.0
buy = 0
sell = 0
df_1['return'] = 1

for i in df_1.index:
    if df_1.shift(1).loc[i, 'trade'] == '' and df_1.loc[i, 'trade'] == 'buy':
        buy = df_1.loc[i, 'Adj Close']
        print('진입일:', i, '구매가:', buy)
    elif df_1.shift(1).loc[i, 'trade'] == 'buy' and df_1.loc[i, 'trade'] == '':
        sell = df_1.loc[i, 'Adj Close']
        rtn = (sell - buy) / buy + 1
        df_1.loc[i, 'return'] = rtn
        print('청산일:', i, '판매가:', sell, '수익률:', round(rtn, 4))
```

```python
acc_rtn = 1.0

for i in df_1.index:
    rtn = df_1.loc[i, 'return']
    acc_rtn *= rtn
    df_1['acc_rtn'] = acc_rtn

print('누적수익률:', round(acc_rtn, 4))
```

```python
def first(data, col, start):
    df = data[~data.isin([np.nan, np.inf, -np.inf]).any(1)]
    df = df.loc[:,[col]]
    df['center'] = df[col].rolling(20).mean()
    df['ub'] = df['center'] + 2 * (df[col].rolling(20).std())
    df['lb'] = df['center'] - 2 * (df[col].rolling(20).std())
    df = df.loc[start :]

    return df
```

```python
df_2 = pd.read_csv('../csv/BND.csv', index_col='Date')
```

```python
df_2 = first(df_2, 'Adj Close', '2010-01-01')
```

```python
def second(data):
    df = data
    df['trade'] = ''
    col = df.columns[0]

    for i in df.index:
        if df.loc[i, col] > df.loc[i, 'ub']:
            df.loc[i, 'trade'] = ''
        elif df.loc[i, col] < df.loc[i, 'lb']:
            df.loc[i, 'trade'] = 'buy'
        elif df.loc[i, col] >= df.loc[i, 'lb'] and df.loc[i, col] <= df.loc[i, 'ub']:
            if df.shift(1).loc[i, 'trade'] == 'buy':
                df.loc[i, 'trade'] = 'buy'
            else:
                df.loc[i, 'trade'] = ''
    
    return df
```

```python
df_2 = second(df_2)
```

```python
def third(data):
    df= data
    df['return'] = 1
    col = df.columns[0]
    
    rtn = 1.0
    buy = 0
    sell = 0

    for i in df.index:
        if df.shift(1).loc[i, 'trade'] == '' and df.loc[i, 'trade'] == 'buy':
            buy = df.loc[i, col]
        elif df.shift(1).loc[i, 'trade'] == 'buy' and df.loc[i, 'trade'] == '':
            sell = df.loc[i, col]
            rtn = (sell - buy) / buy + 1
            df.loc[i, 'return'] = rtn

    acc_rtn = 1.0

    for i in df.index:
        rtn = df.loc[i, 'return']
        acc_rtn *= rtn
        df['acc_rtn'] = acc_rtn

    print('누적수익률:', round(acc_rtn, 3))

    return df
```
