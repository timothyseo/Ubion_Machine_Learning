```python
import pandas as pd

card_df = pd.read_excel('./datasets/default of credit card clients.xls' , index_col=None, header=1)
```

```python
X_features = card_df.drop(columns='default payment next month' , axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

corr = X_features.corr() # 상관관계표
plt.figure(figsize=(14,14)) # plot size
sns.heatmap(corr, annot=True, fmt=".1g") # plot heatmap
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np


labels = card_df['default payment next month']
rf_clf = RandomForestClassifier(n_estimators=300, random_state=126)

scores = cross_val_score(rf_clf,X_features , labels ,  cv=3, scoring='accuracy')
np.mean(scores)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


scaled_features = StandardScaler().fit_transform(X_features)
pay = scaled_features[:,6:12]

pca_2 = PCA(n_components=2)
pca_pay = pca_2.fit_transform(pay)

bill = scaled_features[:,12:18]
pca_bill = pca_2.fit_transform(bill)

cols_bill = ["BILL_AMT"+ str(i) for i in range(1,7)]
cols_pay = ["PAY_"+ str(i) for i in range(2,7)]
cols_drop = cols_bill + cols_pay

df = card_df.drop(columns=cols_drop,axis=1)
drop_df = df.drop(columns="PAY_0",axis=1 )
```

```python
drop_df[['BILL_AMT_1','BILL_AMT_2']] = pca_bill
```

```python
drop_df[['PAY_1','PAY_2']] = pca_pay
```

```python
X_features = drop_df.drop(columns='default payment next month' , axis=1)
```

```python
rf_clf = RandomForestClassifier(n_estimators=300, random_state=126)

scores = cross_val_score(rf_clf,X_features , labels ,  cv=3, scoring='accuracy')
np.mean(scores)
```
