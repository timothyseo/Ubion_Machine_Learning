```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# 가상 데이터 생성
x = data = np.linspace(1, 2, 200) # 시작값 1, 종료값 2, 개수 200
y = x * 4 + np.random.randn(200) * 0.3 # x를 4배로 하고 편차 0.3 정도의 가우시안 잡음 추가

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1, input_dim=1, activation='linear'))
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
model.fit(x, y, batch_size=1, epochs=30)

predict = model.predict(data)

plt.plot(data, predict, 'b', data, y, 'k.') #첫번째 그래프는 파란색 마커로 두번째는 검정색.으로 그린다
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 공부시간 x와 성적 y의 리스트
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

# 그래프
plt.figure(figsize=(8, 5))
plt.scatter(x, y)
plt.show()
```

```python
# 리스트 x, y의 값을 넘파이 배열로 바꿈
x_data = np.array(x)
y_data = np.array(y)

# 기울기 a와 절편 b의 값 초기화
a = 0
b = 0

# setting learning rate
lr = 0.03

# no. of try
epochs = 2023

# gradient decent to go
for i in range(epochs): 
    y_hat = a * x_data + b
    error = y_data - y_hat
    a_diff = -(2/len(x_data)) * sum(x_data * (error))
    b_diff = -(2/len(x_data)) * sum(error)
    mse = np.mean(error**2)
    a = a - lr * a_diff
    b = b - lr * b_diff
    if i % 100 == 0:
        print('epoch = %.f, 기울기 = %.04f, 절편 = %.04f' % (i, a, b))
        print(mse)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# 공부시간 x와 성적 y의 리스트
data = [[2, 0, 81], [4, 4, 93], [6, 2, 91], [8, 3, 97]]
x1 = [i[0] for i in data]
x2 = [i[1] for i in data]
y = [i[2] for i in data]

# 그래프
ax = plt.axes(projection='3d')
ax.set_xlabel('study_hours')
ax.set_ylabel('private_class')
ax.set_zlabel('Score')
ax.dist = 11
ax.scatter(x1, x2, y)
plt.show()
```

```python
# 리스트 x, y의 값을 넘파이 배열로 바꿈
x1_data = np.array(x1)
x2_data = np.array(x2)
y_data = np.array(y)

# 기울기 a와 절편 b의 값 초기화
a1 = 0
a2 = 0
b = 0

# setting learning rate
lr = 0.02

# no. of try
epochs = 2023

# gradient decent to go
for i in range(epochs): 
    y_pred = a1 * x1_data + a2 * x2_data + b
    error = y_data - y_pred
    a1_diff = -(2/len(x1_data)) * sum(x1_data * (error))
    a2_diff = -(2/len(x2_data)) * sum(x2_data * (error))
    b_new = -(2/len(x1_data)) * sum(error)
    mse = np.mean(error**2)
    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_new
    if i % 100 == 0:
        print('epoch = %.f, 기울기1 = %.04f, 기울기2 = %.04f, 절편 = %.04f' % (i, a1, a2, b))
        print(mse)
```
